```python
from langchain_community.llms import Ollama

# Load Ollama with the DeepSeek-R1 model
llm = Ollama(model="deepseek-r1")

# Define a function for structured responses
def generate_response(query):
    response = llm.invoke(query)
    
    # Format the response output
    formatted_response = {
        "query": query,
        "response": response,
        "model": "DeepSeek-R1",
        "provider": "Ollama"
    }
    return formatted_response

# Example Query
context = "Sujeet is a a senior software engineer in Kaplan."
query = f"context: {context} question: What is Sujeet's job title?"
output = generate_response(query)

# Print formatted response
print(output)
```
